```python
# 3. Regression

# We will now run some regression to understand how employee features 
# influence their performance.

y = hiringData.mainPerformanceMetric

# 3.1 Regression with all variables
# Performance regression with all variables - notice that maturity assessment 
# and outside experience are no longer significant once including tenure/age. 

x = hiringData.drop({"mainPerformanceMetric","ID"},axis=1)
model1 = sm.OLS(y,x).fit()

print(model1.summary())

# You can also produce a table with just the coefficients with significance 
# stars using the code below 
print("\n")
print("3.1.2 Table with Significance Stars")
print(summary_col(model1, stars=True))
```

```python
# 3.1 Maturity exercise 
# Observe that maturity has a positive effect when tenure, age and skillTest2 are not included

x_2 = hiringData.drop({"mainPerformanceMetric","ID","Tenure","Age","skillTest2"},axis=1)
model2 = sm.OLS(y,x_2).fit()
print(model2.summary())

# You can also produce a table with just the coefficients with significance 
# stars using the code below 
print("\n")
print("Table with Significance Stars")
print(summary_col(model2, stars=True))
```

```python
# If we include tenure, then maturity should be insignificant 
x_3 = hiringData.drop({"mainPerformanceMetric","ID","Age","skillTest2"},axis=1)
model3 = sm.OLS(y,x_3).fit()
print(model3.summary())

# You can also produce a table with just the coefficients with significance 
# stars using the code below 
print("\n")
print("Table with Significance Stars")
print(summary_col(model3, stars=True))
```

```python
# Same if we include age instead of tenure.
x_4 = hiringData.drop({"mainPerformanceMetric","ID","Tenure","skillTest2"},axis=1)
model4 = sm.OLS(y,x_4).fit()
print(model4.summary())

# You can also produce a table with just the coefficients with significance 
# stars using the code below 
print("\n")
print("Table with Significance Stars")
print(summary_col(model4, stars=True))
```

```python
# 3.2 MBA/team quality exercise: didmba should have a positive effect on performance 
# if we don't include the team quality variable. But note that I cannot include
# outsideexperience either because otherwise MBA is not positive.

x_5 = hiringData.drop({"mainPerformanceMetric","ID","teamQuality","outsideExperience"},axis=1)
model5 = sm.OLS(y,x_5).fit()
print(model5.summary())

# You can also produce a table with just the coefficients with significance 
# stars using the code below 
print("\n")
print("Table with Significance Stars")
print(summary_col(model5, stars=True))
```

```python
# But the effect goes away when we include the team quality variable
x_6 = hiringData.drop({"mainPerformanceMetric","ID","outsideExperience"},axis=1)
model6 = sm.OLS(y,x_6).fit()
print(model6.summary())

# You can also produce a table with just the coefficients with significance 
# stars using the code below 
print("\n")
print("Table with Significance Stars")
print(summary_col(model6, stars=True))
```
